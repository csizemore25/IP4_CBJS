```python
car_sales['Price']=car_sales['Price'].astype('int64')
```

```python
car_sales.info()
```

### Question 6: 

```python
price=np.array(car_sales['Price'])
price
```

```python
total_sales=np.cumsum(price)
car_sales.insert(5, 'Total Sales', total_sales, True)

car_sales.head()
```

### Question 16:

```python
heat_map_corr=car_sales[['Total Sales', 'Price', 'Odometer (KM)']].corr(method='pearson')
print(heat_map_corr)

cols=['Total Sales', 'Price', 'Odometer (KM)']

heat_map=sns.heatmap(heat_map_corr, annot=True, xticklabels=cols, yticklabels=cols)
```

Total Sales and Price: 16% positively correlated - not highly correlated 
Total Sales and Odometer: 32% negatively correlated - not highly correlated 
Price and Odometer: 63$ negatively correlated - fairly highly correlated 

Any same comparison (ex. price to price) is 100% correlated because they are the same values
